```python
afk = pd.read_csv("Прошлые данные - AFKS.csv")
afk_p = list(afk['Цена'][:200])[::-1]
```

```python
p = [tcs_p, gazp_p, agr_p, afk_p]
n = len(p)
r = [[] for i in range(n)]
mu = [0 for i in range(n)]
d = [0 for i in range(n)]
```

```python
afk_p
```

```python
for j in range(n):
  for i in range(len(p[j])):
    p[j][i] = float(p[j][i].replace('.', '').replace(',', '.'))
  # вычисление нетто-доходности
  for i in range(1, len(p[j])):
    r[j].append((p[j][i]/p[j][i-1]) - 1)
  # вычисляем выборочное средние
  mu[j] = sum(r[j])/len(r[j])
  d[j] = sum([(r[j][i] - mu[j]) ** 2 for i in range(len(r[j]))])/len(r[j])
```

```python
print('mu:', mu)
print('d:', d)
```

```python
cov = [[0 for i in range(n)] for i in range(n)]
for j in range(n):
  for i in range(n):
    cov[j][i] = sum([(r[j][k] - mu[j])*(r[i][k] - mu[i]) for k in range(len(r[j]))])/len(r[j])
```

```python
#Выводим матрицу ковариации
for i in range(n):
  print(cov[i])
```

```python
m = (max(mu) + min(mu))/2
```

```python
from sympy import *
k = 10

X = list(symbols(' x1:'+str(n+1)))
LL = list(symbols('l0:'+str(n+3)))
XLL = X.copy()
XLL.extend(LL)

D = sum([sum([X[i]*X[j]*cov[i][j] for i in range(n)]) for j in range(n)])

q1 = sum([X[i]*mu[i] for i in range(n)]) - m

q2 = sum([X[i] for i in range(n)]) - 1

Q = []
for i in range(0,n):
  Q.append(X[i]*LL[i+3])
L = 1*D+LL[1]*q1+LL[2]*q2

for qi in Q:
  L = L - qi
Lx = []
for xi in X:
  Lx.append(L.diff(xi))

LxQqq = Lx.copy()
LxQqq.extend(Q)
LxQqq.append(q2)
LxQqq.append(q1)

#solutions=solve([Lx[0], Lx[1], Lx[2], Q[0], Q[1], Q[2], q2, q1])
solutions=solve(LxQqq)
  #A = np.array[LX,Q,q2,q1]
  #solutions=solve([Lx], X,LL)
```

```python
solutions
```

```python
#Ищем точки потенциального минимум дисперсии
for i in range(len(solutions)):
  solution_list = list(solutions[i].values())[2:]
  if all(x >= 0 for x in solution_list):
    print(solutions[i])
```

```python
#Ищем точки потенциального максимума дисперсии
for i in range(len(solutions)):
  solution_list = list(solutions[i].values())[2:]
  if all(x <= 0 for x in solution_list[:n]) and all(x >= 0 for x in solution_list[n:]):
    print(solutions[i])
```
