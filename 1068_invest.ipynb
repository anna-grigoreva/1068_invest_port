##Проведение расчета выборочных средних, дисперсий и матрицы ковариации

```python
p = [tcs_p, gazp_p, agr_p, afk_p]
n = len(p)
r = [[] for i in range(n)]
mu = [0 for i in range(n)]
d = [0 for i in range(n)]
```

```python
for j in range(n):
  for i in range(len(p[j])):
    p[j][i] = float(p[j][i].replace('.', '').replace(',', '.'))
  # вычисление нетто-доходности
  for i in range(1, len(p[j])):
    r[j].append((p[j][i]/p[j][i-1]) - 1)
  # вычисляем выборочное средние
  mu[j] = sum(r[j])/len(r[j])
  # вычисляем дисперсии
  d[j] = sum([(r[j][i] - mu[j]) ** 2 for i in range(len(r[j]))])/len(r[j])
```

```python
#Выводим выборочные средние и дисперсии каждого актива
print('mu:', mu)
print('d:', d)
```

```python
#Вычисляем ковариации
cov = [[0 for i in range(n)] for i in range(n)]
for j in range(n):
  for i in range(n):
    cov[j][i] = sum([(r[j][k] - mu[j])*(r[i][k] - mu[i]) for k in range(len(r[j]))])/len(r[j])
```

```python
#Выводим матрицу ковариации
for i in range(n):
  print(cov[i])
```

##Переходим к задаче оптимизации с ограничениями в виде неравенств

```python
from sympy import *
k = 10

X = list(symbols(' x1:'+str(n+1)))
LL = list(symbols('l0:'+str(n+2)))
XLL = X.copy()
XLL.extend(LL)

D = sum([sum([X[i]*X[j]*cov[i][j] for i in range(n)]) for j in range(n)])

q1 = sum([X[i] for i in range(n)]) - 1

Q = []
for i in range(0,n):
  Q.append(X[i]*LL[i+2])
L = 1*D+LL[1]*q1

for qi in Q:
  L = L - qi
Lx = []
for xi in X:
  Lx.append(L.diff(xi))

#Список уравнений итоговой системы
LxQ = Lx.copy()
LxQ.extend(Q)
LxQ.append(q1)

solutions=solve(LxQ)
```

```python
#Выводим решения системы
for i in range(len(solutions)):
  print(solutions[i])
```

```python
#Ищем точки потенциального минимум дисперсии
for i in range(len(solutions)):
  solution_list = list(solutions[i].values())[1:]
  if all(x >= 0 for x in solution_list):
    res = list(solutions[i].values())[1:]
    for j in range(len(res[n:])):
      print('x'+str(j+1), ': ', res[j+n], '; ', sep='', end='')
    print()
```

```python
#Ищем точки потенциального максимума дисперсии
for i in range(len(solutions)):
  solution_list = list(solutions[i].values())[1:]
  if all(x <= 0 for x in solution_list[:n]) and all(x >= 0 for x in solution_list[n:]):
    res = list(solutions[i].values())[1:]
    for j in range(len(res[n:])):
      print('x'+str(j+1), ': ', res[j+n], '; ', sep='', end='')
    print()
```
